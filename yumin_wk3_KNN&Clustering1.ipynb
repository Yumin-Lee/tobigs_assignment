```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = load_iris() #iris 데이터 불러오기
```

```python
X = iris.data
y = iris.target
```

```python
X #feature
```

```python
y #label
```

```python
# Train, Test 셋 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=48)
```

```python
# 데이터셋 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# build KNN model (k = 3)
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
# train the model
knn.fit(X_train, y_train)
```

```python
# get the predict value(y_pred) from X_test
y_pred = knn.predict(X_test)
```

```python
# 성능 지표 : accuracy
print('accuracy: ', knn.score(X_test, y_test))
```

# 2. KNN에 GridSearch 적용

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 데이터 정규화를 위한 라이브러리
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# 데이터 스케일링
scaler.fit(X_train)
scaler.fit(X_test)
```

```python
# parameter grid 생성
grid_params = {
    'n_neighbors' : [3,5,11,19],
    'weights' : ['uniform', 'distance'],
    'metric' : ['euclidean', 'manhattan']
}
```

```python
# 그리드서치 적용
gs = GridSearchCV(
    KNeighborsClassifier(),
    grid_params,
    verbose = 1,
    cv = 3,
    n_jobs = -1
    )
```

```python
gs_results = gs.fit(X_train, y_train)
```

# 3. BestParameter & bestScore

```python
# 가장 좋은 결과를 얻는 그리드를 출력
print("Best Parameter: {}".format(gs.best_params_))
```

```python
#grid_search의 최종 스코어를 출력
print("Best Cross-validity Score: {:.3f}".format(gs.best_score_))
```

# 4. Test set Score

```python
#Test 데이터에 대해 grid_search 알고리즘을 적용함
print("Test set Score: {:.3f}".format(gs.score(X_test, y_test)))
```
