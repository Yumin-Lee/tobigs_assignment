```python
import random

class MyDataLoader2:
    def __init__(self, data1, data2):
        # data1과 data2 크기는 같다고 가정합시다
        self.size = len(data1) 
        self.data1 = data1
        self.data2 = data2
        
    # __iter__에서는 나를 반환해주고 __next__에서는 for loop가
    # 돌때마다 인쇄하고 싶은 값을 리턴합니다.
    def __iter__(self): 
        self.index = 0
        return self

    def __next__(self):  # for 루프가 돌 떄마다 인쇄하고 싶은 값 return
        if self.index >= self.size:
            raise StopIteration # size보다 커지면 loop 끝
        n1 = self.data1[self.index]
        n2 = self.data2[self.index]
    
        self.index += 1

        return n1 + n2
       
    def shuffle(self):
        random.shuffle(self.data1)
        random.shuffle(self.data2)

    def make_square(self):
        self.data1 = [x**2 for x in self.data1]
        self.data2 = [x**2 for x in self.data2]
```

```python
# data1과 data2 크기는 같다고 가정합시다
loader = MyDataLoader2([i for i in range(300)], [i for i in range(300)])
loader.make_square() ## data1과 data2를 제곱할 수 있는 method를 완성해주세요!
# 처음 data1은 0,1,4,9,... data2도 마찬가지.
# 셔플해준 후 데이터를 인쇄해주면 -> 두개 더한 값 프린트 해주고.. 
# 와 같은 식으로 커스텀(next에서!)
loader.shuffle()
for i, x in enumerate(loader): # for loop 가 도니까 __iter__
    print("{} : {}".format(i,x)) # i번째 data1과 data2를 더한 값을 리턴하도록 만들어주세요!
```
