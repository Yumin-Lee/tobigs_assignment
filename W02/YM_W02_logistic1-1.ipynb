# 로지스틱 회귀분석

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaled1 = scaler.fit_transform(X_train)
X_strain = pd.DataFrame(scaled1, columns = X_train.columns)
```

```python
scaled2 = scaler.fit_transform(X_test)
X_stest = pd.DataFrame(scaled2, columns=X_test.columns)
```

```python
LogisReg = LogisticRegression()
LogisReg.fit(X_strain, y_train)  #regression fit
```

```python
pred1=LogisReg.predict_proba(X_stest)
p1 = [value[1] for value in pred1]
```

```python
from sklearn.metrics import roc_curve
```

```python
pred_score = p1
fpr, tpr, thresholds = roc_curve(y_test, pred_score)
x = fpr
y = tpr
```

```python
plt.plot(x,y) # 간단하게 ROC 그려볼 수 있음.
plt.xlabel('1-Specificity')
plt.ylabel('Sensitivity')
```
