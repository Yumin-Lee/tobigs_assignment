# 주제 1.
# 유류비 최소화 강화학습

집(출발지)에서 투빅스 세미나 장소(목적지)까지 운전해서 가려고 합니다.

목적지에 가기 전에 1)주유소에 들러 기름 좀 넣고, 2)유민이도 중간에서 태우고, 3)HDMI선도 구입하려 하는데, 이 때 고려해야 할

1. 비용 최소화 (유류비, 통행료)
2. 시간 최소화(교통상황(어디에서 정체되는지), 이동 경로)
3. 앞의 사항에 포함되긴 하지만, 총 이동거리 최소화 

세 가지 사항을 강화학습으로 해결할 수 있을 것이라고 생각했습니다. 

설명의 편의를 위해 여기에서는 경유지 3곳을 각각 주유소1, 2, 3이라고 부를게요

가는 길에는 매우 많은 주유소가 있고, 여기서 3곳의 주유소를 선택한 후 같은 양만큼 주유하는 상황이라고 합니다


## 1. Environment

- 각 주유소 별로
    - 출발지로부터의 이동거리 (경로를 따라서 이동한 거리)
    - 도착지까지의 이동거리 (경로를 따라서 이동한 거리)
- 모든 주유소 사이 각각의 이동거리
    
 ## 2. State
- 주요소 별 유가

- 교통상황
    - 출발지로부터의 주행 시간 (거리와는 다른 관점)
    - 도착지까지의 주행 시간 (거리와는 다른 관점)
  
## 3. Action
- 각 step별로 주유소 선정

## 4. Reward 
각 step별로
- Reward = - {(유류비 + 통행료) + ( (1/시속) * 이동거리 )}

- 0에 가까울수록 좋은 값
    - 각 요소들은 모두 양수이므로 reward는 음수값으로 반환됨
    - 이 값의 절대값이 작을수록 좋은 값
    - 목표: 유류비, 통행료, 이동거리는 최소화
    - 목표: 시속은 최대화

# 주제 2. 
#  가장 효율적인 방법으로 놀이기구 즐기기

설명: 놀이동산에서 놀이기구를 탈 때, 고려해야 하는 여러 사항들을 강화학습을 통해 가장 효율적으로 고려할 수 있을 것이라고 생각했습니다.

여러 조건들을 고려해서 주어진 시간 안에 
1. 가장 많은 놀이기구 타기 
2. 가장 최소화된 동선으로 놀이기구 타기

등의 문제를 해결하려고 합니다.

## 1. Environment
- 놀이기구들 사이를 이동하는 데 걸리는 시간
- 각 놀이기구 별로
    - 탑승 시간
    - 한 번에 탑승 가능한 인원 수
    
## 2. State
- 각 놀이기구 별 대기 인원

## 3. Action
- 다음 탑승할 놀이기구 선정

## 4. Reward
- Reward = - {1/(탑승 인원) + (탑승 시간, 대기인원, 놀이기구간 이동거리)}
    - 0에 가까울수록 좋은 값
    - 요소는 모두 음수이므로 절대값이 작을수록 좋은 값
    - 목표: 탑승 시간, 대기인원, 놀이기구간 이동거리 최소화
    - 목표: 탑승 인원 최대화
